```python
# Definir una función para eliminar las filas que no contengan 'TULUÁ' en la columna 'MUNICIPIO'
def filtrar_por_municipio(data):
    data_filtrado = data[data['MUNICIPIO'] == 'TULUÁ']
    return data_filtrado


# Filtrar los DataFrames por el valor 'TULUÁ' en la columna 'MUNICIPIO'
data1_filtrado_amena_actualAD = filtrar_por_municipio(data_amenazas_11)
data2_filtrado_homi_actualAD = filtrar_por_municipio(data_homicidio_1)
data3_filtrado_extor_actualAD = filtrar_por_municipio(data_extorcion_9)
data4_filtrado_extor_actualAD = filtrar_por_municipio(data_extorsion_5)
data5_filtrado_extor_gustavoAD = filtrar_por_municipio(data_extorsion_2017)
data6_filtrado_extor_gustavoAD = filtrar_por_municipio(data_extorsion_2018)
data7_filtrado_homi_gustavoAD = filtrar_por_municipio(data_homicidios_2018)
data8_filtrado_amena_gustavoAD = filtrar_por_municipio(data_Amenazas_2019)
```
